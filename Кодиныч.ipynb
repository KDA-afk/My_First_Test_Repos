```python
import pandas as pd     # для чтения и записи табличных файлов
import numpy as np      # для математической обработки данных
import matplotlib.pyplot as plt # для построения графиков
from bokeh.plotting import figure, output_file, show
from bokeh.io import output_notebook

# Чтение CSV-файла
url = "https://drive.google.com/uc?export=download&id=1XFViS21mLmAnVUqzVsMYgxZUPw6sYn00"
data = pd.read_csv(url)
# вывод заголовка таблицы и первых пяти строк (сэмплов)
head = data.head()
print(head)

data['Solar_power_MW'] = data['Solar_power_kW'] / 1000
data['Difference'] = (data['Wind_power_MW'] + data['Solar_power_MW']) - data['Consumption_MW']

# Проверка измененмй таблицы
head = data.head()
print(head)

#подсчёт всех значений
Itog = data['Difference'].sum()

# округление значений до 4 знаков после запятой
Itog = Itog.round(4)
#axes = data['Difference'].plot(figsize=(20, 7), subplots = True)

# создание нового файла для удобства
data_gr = pd.DataFrame(data['Difference'])

sol = data_gr.head()
print(sol)
print('можно продать',Itog , 'MW за весь период')

# для работы с интерактивным графиком
output_notebook()

# Берем только ряд с значением разницы из нового датафрэйма
# iloc - выделяет из таблицы указанный фрагмент, в данном случае все строки и только первый столбец
Difference = data_gr.iloc[: , 0].values
Wind = data.iloc[ : , 0].values
solar = data.iloc[ : , 3].values
consum = data.iloc[ : , 2].values
#print(Difference)
#print(solar)

#Размеры графика
p1 = figure(width= 800, height= 400)
p2 = figure(width= 800, height= 400)
p3 = figure(width= 800, height= 400)
p = figure(width= 800, height= 400)

#print(len(Wind))
#print(Wind)

#создание графика
p1.line(np.arange(len(Wind)), Wind , line_width= 2, color='green')
p2.line(np.arange(len(Wind)), solar , line_width= 2, color='orange')
p3.line(np.arange(len(Wind)), consum , line_width= 2, color='blue')
p.line(np.arange(len(Difference)), Difference, line_width = 2, color='red')

#отрисовка графика
show(p1)
show(p2)
show(p3)
show(p)
```
